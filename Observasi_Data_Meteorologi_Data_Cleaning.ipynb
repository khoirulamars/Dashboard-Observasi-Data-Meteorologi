## Data AWS

### Informasi Dataset AWS

```
'Date': Tanggal pengukuran.
'Time': Waktu pengukuran.
'Temp Out': Temperatur udara di luar.
'Hi Temp': Temperatur tertinggi yang tercatat.
'Low Temp': Temperatur terendah yang tercatat.
'Out Hum': Kelembaban udara di luar.
'Dew Pt.': Titik embun.
'Wind Speed': Kecepatan angin.
'Wind Dir': Arah angin.
'Wind Run': Jarak yang ditempuh oleh angin dalam satu periode waktu tertentu.
'Hi Speed': Kecepatan angin tertinggi yang tercatat.
'Hi Dir': Arah angin saat kecepatan tertinggi terjadi.
'Wind Chill': Sensasi dingin yang dirasakan oleh tubuh karena angin dan suhu udara.
'Heat Index': Indeks suhu yang menggambarkan sensasi panas yang dirasakan oleh tubuh.
'THW Index': Indeks panas yang memperhitungkan suhu udara, kelembaban, dan angin.
'Bar': Tekanan udara atmosfer.
'Rain': Jumlah curah hujan yang tercatat.
'Rain Rate': Intensitas curah hujan dalam satuan waktu tertentu.
'Heat D-D': Derajat hari (Degree Days) untuk pendinginan.
'Cool D-D': Derajat hari (Degree Days) untuk pemanasan.
'In Temp': Temperatur di dalam ruangan.
'In Hum': Kelembaban di dalam ruangan.
'In Dew': Titik embun di dalam ruangan.
'In Heat': Panas di dalam ruangan.
'In EMC': Kandungan air dalam material.
'In Air Density': Kepadatan udara di dalam ruangan.
'Wind Samp': Sampel kecepatan angin.
'Wind Tx': Transmisi data kecepatan angin.
'ISS Recept': Penerima data stasiun cuaca.
'Arc. Int.': Intensitas radiasi sinar matahari yang diterima.
```

### AWS - Data Cleaning

#### Data Sebelum Pengecetan

````python
aws_bef = pd.read_excel('/content/drive/MyDrive/Satu Data/Data KKN/AWS 29 Desember - 1 Januari 2024.xlsx')
aws_bef.head()
````

````python
aws_bef = aws_bef.iloc[3:430].reset_index(drop=True)
````

````python
aws_bef.head()
````

````python
# Mengecek kolom yang terdapat nilai '---'
def check_for_dashes(column):
    return '---' in column.values

columns_with_dashes = aws_bef.apply(check_for_dashes)
columns_with_dashes[columns_with_dashes].index.tolist()
````

````python
aws_bef.replace('---', np.nan, inplace=True)
````

````python
aws_bef.isnull().sum()
````

````python
aws_bef.info()
````

````python
# Konversi tipe data Date and Time
aws_bef['Date'] = pd.to_datetime(aws_bef['Date'])
aws_bef['Time'] = pd.to_datetime(aws_bef['Time'], format='%H:%M:%S').dt.time

aws_bef['DateTime'] = aws_bef.apply(lambda row: pd.datetime.combine(row['Date'], row['Time']), axis=1)
````

````python
# Ubah tipe data
columns_bef = ['Temp Out', 'Hi Temp', 'Low Temp', 'Dew Pt.', 'Wind Chill', 'Heat Index', 'THW Index', 'Heat D-D', 'Cool D-D']

for i in columns_bef:
  aws_bef[i] = aws_bef[i].astype(float)
````

````python
aws_bef.fillna(aws_bef.mean(), inplace=True)
````

````python
aws_bef.head(10)
````

````python
aws_bef['Wind Dir'].unique()
````

````python
aws_bef['Hi Dir'].unique()
````

````python
aws_bef.to_csv('dataAWS_bef.csv')
````

#### Data Setelah Pengecetan

````python
aws_aft = pd.read_excel('/content/drive/MyDrive/Satu Data/Data KKN/AWS 8 - 11 Januari 2024.xlsx')
aws_aft.head()
````

````python
def check_for_dashes(column):
    return '---' in column.values

columns_with_dashes = aws_aft.apply(check_for_dashes)
columns_with_dashes[columns_with_dashes].index.tolist()
````

````python
aws_aft.replace('---', np.nan, inplace=True)
````

````python
aws_aft.isnull().sum()
````

````python
aws_aft.info()
````

````python
# Mengubah format Date and Time
aws_aft['Date'] = pd.to_datetime(aws_aft['Date'], format='%d/%m/%y')
aws_aft['DateTime'] = aws_aft['Date'].astype(str) + ' ' + aws_aft['Time']

aws_aft['DateTime'] = pd.to_datetime(aws_aft['DateTime'], format='%Y-%m-%d %H:%M')
````

````python
# Ubah tipe data
columns_aft = ['Temp Out', 'Hi Temp','Out Hum','Low Temp', 'Dew Pt.', 'Wind Chill', 'Heat Index', 'THW Index', 'Heat D-D', 'Cool D-D']

for i in columns_aft:
  aws_aft[i] = aws_aft[i].astype(float)
````

````python
aws_aft.info()
````

````python
aws_aft.fillna(aws_aft.mean(), inplace=True)
````

````python
aws_aft.isnull().sum()
````

````python
aws_aft.head(10)
````

````python
aws_aft.to_csv('dataAWS_aft.csv')
````
